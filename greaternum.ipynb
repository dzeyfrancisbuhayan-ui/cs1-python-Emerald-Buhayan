```python
num1 = int(input("Enter the first number: "))
num2 = int(input("Enter the second number: "))
num3 = int(input("Enter the third number: "))

print(f"You entered: {num1}, {num2}, {num3}")

# Compare the numbers to find the biggest one
biggest_number = num1
if num2 > biggest_number:
    biggest_number = num2
if num3 > biggest_number:
    biggest_number = num3

# Print out the biggest number
print(f"The biggest number is: {biggest_number}")
```
