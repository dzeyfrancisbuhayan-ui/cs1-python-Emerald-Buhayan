```python
score_input = input("Enter student's score: ")
score = int(score_input)

if 90 <= score <= 100:
    grade = 'A'
elif 80 <= score <= 89:
    grade = 'B'
elif 70 <= score <= 79:
    grade = 'C'
else:
    grade = 'F'

print(f"\n Grading Result ")
print(f"Score: {score}")
print(f"Letter Grade: {grade}")
```
